```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#purchase_data.describe()
count=len(purchase_data['SN'].unique())
count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data.head()
uniqueItem=len(purchase_data['Item Name'].unique())
ave=purchase_data['Price'].mean()
totalpurchase=sum(purchase_data['Price'])
numberpurchase=sum(purchase_data['Purchase ID'])

analysis_df = pd.DataFrame([{"Number of Unique Items": uniqueItem, "Average Price": ave,
                                      "Number of Purchases": numberpurchase, "Total Revenue": totalpurchase}])
analysis_df["Average Price"] =analysis_df["Average Price"].map("${:,.2f}".format)
analysis_df["Total Revenue"] =analysis_df["Total Revenue"].map("${:,.2f}".format)
analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
genders=purchase_data[['SN','Gender']].drop_duplicates()
genders

gender_count=genders['Gender'].value_counts()
gender_count

totals = [gender_count[0],gender_count[1],gender_count[2]]
percents = [round((gender_count[0]/count)*100,2),round((gender_count[1]/count)*100,2),round((gender_count[2]/count)*100,2)]


genders = pd.DataFrame({ "Percentage of Players": percents,"Total Count": totals
})
genders.index = (["Male", "Female", "Other / Non-Disclosed"])
genders
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_df=purchase_data.groupby(['Gender'])

purchase_count=gender_df['SN'].count()
purchase_ave=gender_df['Price'].mean()
purchase_total=gender_df['Price'].sum()

#group duplicates for per person ave
duplicates = purchase_data.drop_duplicates(subset='SN', keep="first")
grouped_duplicates = duplicates.groupby(["Gender"])
per_person=(gender_df["Price"].sum() / grouped_duplicates["SN"].count())

gender_table=pd.DataFrame({'Purchase Count':purchase_count,
                           'Average Purchase Price':purchase_ave,
                           'Total Purchase Value':purchase_total,
                           'Average Per Person':per_person})

gender_table['Average Purchase Price']=gender_table['Average Purchase Price'].map("${:,.2f}".format)
gender_table['Total Purchase Value']=gender_table['Total Purchase Value'].map("${:,.2f}".format)
gender_table['Average Per Person']=gender_table['Average Per Person'].map("${:,.2f}".format)
gender_table
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#make bins
bins=[0,9.9,14.90,19.90,24.90,29.90,34.90,39.90,9999]
names=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

age_df = purchase_data
age_df['Age Groups'] = pd.cut(age_df['Age'], bins, labels = names)
age_df.head()

#group by bins
grouped_df=age_df.groupby('Age Groups')

bin_count=grouped_df['SN'].count()
count_total=purchase_data['SN'].count()
percent=(bin_count/count_total)*100
percent

percentage_df=pd.DataFrame({'Percentage of Players':percent,'Total Count':bin_count})
percentage_df
percentage_df["Percentage of Players"] = percentage_df["Percentage of Players"].map("{:.2f}%".format)
percentage_df

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

#count grouped bins for calculations
bin_count = grouped_df["Age"].count()
bin_ave = grouped_df["Price"].mean()
bin_total = grouped_df["Price"].sum()

#group duplicate SN 
bin_duplicate = purchase_data.drop_duplicates(subset='SN', keep="first")
bin_duplicate["Age Groups"] = pd.cut(bin_duplicate["Age"], bins, labels=names)
bin_duplicate = bin_duplicate.groupby(["Age Groups"])

perPerson = (grouped_df["Price"].sum() / bin_duplicate["SN"].count())
perPerson

#create new df
purch_demo = pd.DataFrame({"Purchase Count": bin_count,
                         "Average Purchase Price": bin_ave,
                         "Total Purchase Value": bin_total,
                         "Average Per Person": perPerson})
purch_demo

#format accordingly
purch_demo['Average Purchase Price']=purch_demo['Average Purchase Price'].map('${:.2f}'.format)
purch_demo['Total Purchase Value']=purch_demo['Total Purchase Value'].map('${:.2f}'.format)
purch_demo['Average Per Person']=purch_demo['Average Per Person'].map('${:.2f}'.format)

purch_demo
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by SN now
groupSN=purchase_data.groupby(['SN'])

sn_count=groupSN['Item ID'].count()
sn_total=groupSN['Price'].sum()
sn_ave=(sn_total/sn_count)

sn_demo=pd.DataFrame({'Purchase Count':sn_count,'Average Purchase Price':sn_ave,'Total Purchase Value':sn_total})
sn_demo

sn_demo["Average Purchase Price"] = sn_demo["Average Purchase Price"].map("${:.2f}".format)
sn_demo["Total Purchase Value"] = sn_demo["Total Purchase Value"].map("${:.2f}".format)
sn_demo=sn_demo.sort_values('Total Purchase Value', ascending=False)
sn_demo
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
groupItem = purchase_data.groupby(["Item ID", "Item Name"])
#same format as above but with "Item"
item_count=groupItem['SN'].count()
item_sum=groupItem['Price'].sum()
item_aveP=(item_sum/item_count)
item_total=(item_aveP*item_count)
#item_price=groupItem['Price']

item_demo=pd.DataFrame({'Purchase Count':item_count, 'Item Price':item_aveP,'Total Purchase Value':item_total})
item_demo

item_demo['Item Price']=item_demo['Item Price'].map('${:.2f}'.format)
item_demo['Total Purchase Value']=item_demo['Total Purchase Value'].map('${:.2f}'.format)
item_demo=item_demo.sort_values('Purchase Count', ascending=False)
item_demo
```
